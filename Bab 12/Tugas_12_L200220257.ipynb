```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
grocery = pd.read_csv('GroceryStoreDataSet.csv')
grocery.head()
```

```python
records = []
for i in range(0, grocery.shape[0]):
    records.append([str(item).strip() for item in grocery.values[i, 0].split(',')])
print(records)
```

```python
association_rules = apriori(records, min_support = 0.003, min_confidence = 0.2, min_left = 3, min_length = 2)
association_results = list(association_rules)
for i in association_results:
    print(i, "\n")
```
