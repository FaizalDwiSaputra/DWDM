```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
store_data = pd.read_csv('store_data.csv')
store_data.head()
```

```python
records = []
for i in range(0, store_data.shape[0]):
    records.append([str(store_data.values[i,j]) for j in range(0, store_data.shape[1])])
print(records)
```

```python
association_rules = apriori(records, min_support = 0.0045, min_confidence = 0.2, min_left = 3, min_length = 2)
association_results = list(association_rules)
for i in association_results:
    print(i, "\n")
```
