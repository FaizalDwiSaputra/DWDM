```python
from sklearn.cluster import KMeans

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
# membuat dataframe
x = pd.read_csv('dwdmclustering.csv') 
```

```python
# memilih fitur untuk clustering
x= x.filter(["B.INDONESIA","B.INGGRIS"], axis=1)
```

```python
# normalisasi data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
# inisialisasi dan penerapan K-Means
kmeans = KMeans(n_clusters=2, random_state=0)
x['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# menampilkan hasil clustering
print("Centroid Cluster:")
print(kmeans.cluster_centers_)
print("\nData dengan Cluster:")
print(x)
```

```python
# visualisasi hasil clustering
plt.figure(figsize=(6, 4))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=x['Cluster'], cmap='viridis', label='Data Points')
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.xlabel('B.Indonesia (scaled)')
plt.ylabel('B.Inggris (scaled)')
plt.title('Clustering K-Means')
plt.legend()
plt.show()
```
