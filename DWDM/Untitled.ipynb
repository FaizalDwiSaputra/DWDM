```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
# Menampilkan seluruh nilai kolom 'genres'
movies['status'].unique()
```

```python
movies['status'] = movies['status'].map({'Realesed':0, 'Post Production':1, 'Rumored':2})
```

```python
# Pilih fitur dan target
X = movies.drop(columns=['index','original_language', 'original_title', 'overview', 'production_companies', 'production_countries', 'spoken_languages', 'title', '' 'homepage', 'id', 'keywords', 'tagline', 'release_date', 'director', 'cast', 'crew', 'genres'])  # Ganti 'target_column' dengan nama kolom target
y = movies['status']
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

print('The shape of X_train is : {}'.format(x_train.shape))
print('The shape of X_test is : {}'.format(x_test.shape))
print('The shape of y_train is : {}'.format(y_train.shape))
print('The shape of y_test is : {}'.format(y_test.shape))
```

```python
model = GaussianNB()
model.fit(x_train, y_train)
```
