```python
train_data['Title'] = train_data['Name'].str.extract(
    '([A-Za-z]+)\.', expand=False)
train_data = train_data.drop(columns='Name')
train_data.head()
```

```python
train_data['Title'].value_counts().plot(kind='bar')
```

```python
train_data['Title'] = train_data['Title'].replace('Mr', '0')
train_data['Title'] = train_data['Title'].replace('Miss', '1')
train_data['Title'] = train_data['Title'].replace('Mrs', '2')
train_data['Title'] = train_data['Title'].replace('Others', '4')
```
