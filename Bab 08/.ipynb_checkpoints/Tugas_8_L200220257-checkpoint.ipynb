```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
breast = pd.read_csv('Breast-Cancer.csv')
```

```python
breast.drop(columns='Age', inplace=True)
breast.isnull().sum()
```

```python
import missingno as msno
msno.bar(breast, figsize=(8,6), color='skyblue')
plt.show()
```

```python
g = sns.relplot(x='BMI', y='Glucose', data=breast, hue='Classification', style='Classification')
g.fig.set_size_inches(10,5)
plt.show()
```

```python
numeric_columns = breast.select_dtypes(include = ["float64", "int64"]).columns
correlation_matrix = breast[numeric_columns].corr()
```

```python
sns.heatmap(correlation_matrix.corr(), annot = True, cmap = "Greens")
plt.show()
```

```python
x = breast.drop(['Classification','Insulin', 'HOMA', 'Adiponectin', 'MCP.1'], axis=1)
y = breast['Classification']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)

print('The shape of X_train is : {}'.format(x_train.shape))
print('The shape of X_test is : {}'.format(x_test.shape))
print('The shape of y_train is : {}'.format(y_train.shape))
print('The shape of y_test is : {}'.format(y_test.shape))
```

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
pred_test = model.predict(x_test)
```

```python
pred_train = model.predict(x_train)
cm = confusion_matrix(y_train, pred_train)

# Confusion matrix
print('Confusion matrix Naive Bayes\n', cm)
print('')

# Akurasi
print('Akurasi pada saat training: {}' .format(accuracy_score(y_train, pred_train)))
```

```python
cm = confusion_matrix(y_test, pred_test)
accuracy = accuracy_score(y_test, pred_test)
precision = precision_score(y_test, pred_test, average='micro')
recall = recall_score(y_test, pred_test, average='micro')
f1 = f1_score(y_test, pred_test, average='micro')

print('Confusion matrix Naive Bayes\n', cm)
print('')
print('Akurasi pada data test: %.3f' %accuracy)
print('precision: %.3f' %precision)
print('recall : %.3f' %recall)
print('f1-score : %.3f' %f1)
```
